```python
df.drop(['Investments (₹)','Savings (₹)','EMI/Loans (₹)','Dining & Entertainment (₹)'],axis='columns',inplace=True)
categorical_col=[]
for column in df.columns:
   if df[column].dtype == 'object' and len(df[column].unique())<=50:
     categorical_col.append(column)
     df['Healthcare (₹)']=df.Attrition.astype('category').cat.codes
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
for column in categorical_col:
  df[column]=label.fit_transform(df[column])
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('Healthcare (₹)',axis=1)
y=df['Healthcare (₹)']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
if 'Healthcare (₹)' in categorical_col:
    categorical_col.remove('Healthcare (₹)')
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
def print_score(clf,X_train,X_test,y_train,y_test,train=True):
  if train:
    pred=clf.predict(X_train)
    clf_report=pd.DataFrame(classification_report(y_train,pred,output_dict=True))
    print("Train Result\n==========================")
    print(f"Accuracy Score:{accuracy_score(y_train,pred)*100:.2f}%")
    print("--------------------------------------------------------")
    print(f"Classification Report:\n{clf_report}")
    print("----------------------------------------------------------")
    print(f"Confusion Metrix:\n{confusion_matrix(y_train,pred)}")

  elif train==False:
     pred=clf.predict(X_test)
     clf_report=pd.DataFrame(classification_report(y_test,pred,output_dict=True))
     print("Train Result\n==========================")
     print(f"Accuracy Score:{accuracy_score(y_test,pred)*100:.2f}%")
     print("--------------------------------------------------------")
     print(f"Classification Report:\n{clf_report}")
     print("----------------------------------------------------------")
     print(f"Confusion Metrix:\n{confusion_matrix(y_test,pred)}")


```

```python
print(x_train.dtypes)
print(x_test.dtypes)
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
for col in x_train.select_dtypes(include='object').columns:
    x_train[col] = label.fit_transform(x_train[col])
    x_test[col] = label.transform(x_test[col])
```

```python
from sklearn.tree import DecisionTreeClassifier
x_train['Month'] = pd.to_numeric(x_train['Month'], errors='coerce')
x_test['Month'] = pd.to_numeric(x_test['Month'], errors='coerce')
tree_clf=DecisionTreeClassifier(random_state=42)
tree_clf.fit(x_train,y_train)
print_score(tree_clf,x_train,x_test,y_train,y_test,train=True)
print_score(tree_clf,x_train,x_test,y_train,y_test,train=False)
```
